## Create some Monte Carlo simulations

### Estimate $\pi$ geometrically

```julia
steps = 100000
n = 0
hits = 0
while n < steps
    n += 1
    r = rand(2)
    if r[1]^2 + r[2]^2 < 1
        hits +=1
    end
end
println("Number of hits: $hits")
pi_est = 4 * hits / steps 
accuracy = abs(pi_est/3.14159 - 1)
println("Estimate for pi: $pi_est, accuracy: $accuracy")
```

### 2d Ising Model

Model of a ferromagnet with a phase transition (unordered -> ordered) at critical temperature T_c = 2.269

$ H = -J \sum_{<i,j>} S_i S_j + h \sum_i S_i$, the first sum is over nearest neighbors, 
Spins $S_i = \{-1, +1\}$

```julia
function total_energy(spin, j, h)
    nx, ny = size(spin)
    external_e = sum(spin) * h
    interaction_e = 0
    for i in 1:nx, j in 1:ny
        j < ny ? spin_right = spin[i, j+1] : spin_right = spin[i, 1] # periodic boundary conditions
        i < nx ? spin_up = spin[i+1, j] : spin_up = spin[1, j] 
        interaction_e += spin[i, j] * (spin_right + spin_up)
    end
    return external_e - j*interaction_e
end
```

```julia
function partial_energy(x, y, spin, j, h)
    nx, ny = size(spin)
    x > 1 ? x1 = x-1 : x1 = nx 
    x < nx ? x2 = x+1 : x2 = 1 
    y > 1 ? y1 = y-1 : y1 = ny
    y < ny ? y2 = y+1 : y2 = 1 
    spin_neighbors = spin[x1, y] + spin[x2, y] + spin[x, y1] + spin[x, y2]
    partial_e = -j* spin[x, y] * spin_neighbors + h*spin[x, y]
    return partial_e
end
```

```julia
nx = 100
ny = 100
spin = rand((1, -1), nx, ny)
steps = 10000000
intermediate = ceil(steps / 20)
n = 0
hits = 0
temperature = 2.27  # T_c = 2.269j
beta = 1 / temperature
h = 0 # external magnetic field
j = 1 # coupling
while n < steps
    n += 1
    x = rand((1:nx))
    y = rand((1:ny))
    e_old = partial_energy(x, y, spin, j, h)
    e_new = - e_old
    delta_e = e_new - e_old
    if exp(-beta*delta_e) > rand()
        hits += 1
        spin[x, y] = - spin[x, y]
    end
    if n % intermediate == 0
        e_total = total_energy(spin, j, h) / (nx*ny)
        s_mean = sum(spin) / (nx*ny)
        println("hits: $hits / $n; total energy: $e_total, magnetization: $s_mean")
    end
end
```

### Simple Liquid in 3d

Model of a simple atomic liquid (such as liquid Argon) using a Lennard-Jones potential $V(r) = \epsilon \ (\sigma^{12} / r^{12}  -  \sigma^6 / r^6) $

The simulation uses periodic boundary conditions (PBC), distances in units of sigma

```julia
function hexagonal_packing(box, num_atoms)
    nx, ny, nz = box
    pos = Array{Float64, 2}(undef, num_atoms, 3)
    n = 0
    for i in 1:nx, j in 1:ny, k in 1:nz
        n += 1
        if n y= num_atoms
            m = k % 2
            hex_point = (i+(j+m)/2, (3j+m)/sqrt(12), k*sqrt(2/3))
            pos[n, :] = hex_point
        end
    end
    return pos
end
```

```julia
function square_packing(box, num_atoms)
    nx, ny, nz = box
    pos = zeros(num_atoms, 3)
    n = 0
    for i in 1:nx, j in 1:ny, k in 1:nz
        n += 1
        if n <= num_atoms
            pos[n, :] = [i-0.5, j-0.5, k-0.5]
        end
    end
    return pos
end
```

```julia
function total_energy(pos, box, epsilon)
    num_atoms = size(pos)[1]
    e_total = 0
    for m in 1:num_atoms, n in (m+1):num_atoms
        dr = pos[m, :] - pos[n, :] - round.(pos[m, :] - pos[n, :]./ box, digits=0) .* box # PBC
        r2 = sum(dr.*dr)
        r6 = r2^3
        r12 = r6^2
        e_pair = epsilon*(1.0/r12 - 1.0/r6)
        e_total += e_pair
    end
    return e_total
end
        
```

```julia
function partial_energy(n_i, pos_i, pos, box, epsilon)
    num_atoms = size(pos)[1]
    e_partial = 0
    m_range = deleteat!(Array(1:num_atoms), n_i) 
    for m in m_range
        dr = pos[m, :] - pos_i - round.((pos[m, :] - pos_i)./box, digits=0) .* box # PBC
        r2 = sum(dr.*dr)
        r6 = r2^3
        r12 = r6^2
        e_pair = epsilon*(1.0/r12 - 1.0/r6)
        e_partial += e_pair
    end
    return e_partial
end
  
```

```julia
function metropolis_step(pos, box, epsilon, delta_r, beta)
    num_atoms = size(pos)[1]
    n_i = rand(1:num_atoms)
    pos_old = pos[n_i, :]
    pos_new = pos[n_i, :] + randn(3) * delta_r
    e_old = partial_energy(n_i, pos_old, pos, box, epsilon)
    e_new = partial_energy(n_i, pos_new, pos, box, epsilon)
    e_diff = e_new -  e_old
    if exp(-beta*e_diff) > rand()
        success = 1
        pos[n_i, :] = pos_new
    else
        success = 0
    end
    return pos, success
end
```

```julia
## test partial energy function
box = [5.0, 5.0, 5.0]
pos = zeros(2, 3)
pos[2,1] = 1
n_i = 2
epsilon=1.0
r = 0.98:0.01:2.0
e_r = zeros(0)
for mr in r
    pos_i = pos[2, :] * mr
    e_old  = partial_energy(n_i, pos_i, pos, box, epsilon)
    append!(e_r, e_old)
end

plot(r, e_r, 
    xlabel="r", 
    ylabel="V(r)",
    title="Lennard-Jones Pair Potential",
    legend = false)
```

```julia
epsilon = 1.0
temperature = 1.0
beta = 1 / temperature
delta_r = 0.1
box =[5.0, 5.0, 5.0]
num_atoms = 125
mc_steps = 1000000
intermediate = ceil(mc_steps / 20)
step = 0
hits = 0
pos_init = pos = square_packing(box, num_atoms)

while step < mc_steps
    step += 1
    pos, success = metropolis_step(pos, box, epsilon, delta_r, beta)
    hits += success
    if step % intermediate == 0
        e_total = total_energy(pos, box, epsilon)
        println("hits: $hits / $step; total energy: $e_total")
    end
end
```

```julia
for n in num_atoms
    pos[n, :] = pos[n,:] - round.((pos[n, :] - box/2)./box , digits=0) .* box # enforce PBC on atoms in box
end
plt3d= Plots.plot(pos[:, 1],pos[:, 2], pos[:, 3],
     seriestype=:scatter, markersize = 12, legend=false)
display(plt3d)
```
